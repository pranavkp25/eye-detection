```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('pranav.jpg')
img
```

```python
plt.imshow(img)
```

```python
img_conv=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
plt.imshow(img_conv)
```

```python
eye_data=cv2.CascadeClassifier('haarcascade_eye.xml')
eye_data
```

```python
eyes=eye_data.detectMultiScale(img_conv,minNeighbors=15)
eyes
```

```python
for (x,y,w,h) in eyes:
    cv2.rectangle(img_conv,(x,y),(x+w,y+h),(0,255,0),3)
```

```python
plt.imshow(img_conv)
```

```python
video=cv2.VideoCapture(0)
while True:
    success,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    eyes=eye_data.detectMultiScale(img1)
    for (x,y,w,h) in eyes:
        cv2.rectangle(img1,(x,y),(x+w,y+h),(255,0,0),3)
        cv2.imshow("Video",img1)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
